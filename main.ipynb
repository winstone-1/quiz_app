```python
import random
import json


print('CLI Quiz Game')
print('-' * 30)
print('Select a category:')
print('  1. Math')
print('  2. General Knowledge')
print('  3. History')

categories = {"1": "Math", "2": "General Knowledge", "3": "History"}

cat_input = input('Enter 1, 2 or 3: ').strip()
category = categories.get(cat_input)

if not category:
    print('Invalid choice, defaulting to General Knowledge')
    category = 'General Knowledge'


try:
    with open('questions.json', 'r') as file:
        all_questions = json.load(file)
except FileNotFoundError:
    print('Error: questions.json not found')
    all_questions = []


questions = [q for q in all_questions if q.get('category') == category]
random.shuffle(questions)


score = 0
missed = []

print(f'\nCategory: {category}')
print('-' * 30)

for i, q in enumerate(questions, start=1):
    print(f'\nQ{i}: {q["question"]}')
    for key, val in q['choices'].items():
        print(f'  {key}. {val}')

    try:
        answer = input('Your answer (A/B/C/D): ').strip().upper()
        if answer not in ['A', 'B', 'C', 'D']:
            raise ValueError('Invalid choice — must be A, B, C or D')

    except ValueError as ex:
        print(f'{ex}\n')
        missed.append(q)
        continue

    if answer == q['answer']:
        print('Correct!\n')
        score += 1
    else:
        print(f'Wrong. The answer was {q["answer"]}. {q["choices"][q["answer"]]}\n')
        missed.append(q)


total = len(questions)
result = score / total * 100

print('-' * 30)
print('You are done with the quiz!!')
print(f'Score: {score}/{total}  ({result:.1f}%)')

if result == 100:
    print('Perfect Score!')
elif result >= 80:
    print('Great job')
elif result >= 60:
    print('Fair job')
elif result >= 40:
    print('Do better next time')
else:
    print('Keep studying!')

if missed:
    print('\nQuestions you missed:')
    for q in missed:
        print(f'  Q: {q["question"]}')
        print(f'  A: {q["answer"]}. {q["choices"][q["answer"]]}\n')
```
